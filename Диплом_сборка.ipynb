```python
df = pd.read_excel('US_Regional_Sales_Data.xlsx', sheet_name = 'Sales Orders Sheet')
df.head()
```

```python
df_customer = pd.read_excel('US_Regional_Sales_Data.xlsx', sheet_name = 'Customers Sheet')
df_customer.head()
```

```python
df_products = pd.read_excel('US_Regional_Sales_Data.xlsx', sheet_name = 'Products Sheet')
df_products.head()
```

```python
df_salesTeam = pd.read_excel('US_Regional_Sales_Data.xlsx', sheet_name = 'Sales Team Sheet')
df_salesTeam.head()
```

```python
df_merged = pd.merge(df, df_customer, on='_CustomerID', how='left')
```

```python
df_merged1 = pd.merge(df_merged, df_salesTeam, on='_SalesTeamID', how='left')
```

```python
df_merged2 = pd.merge(df_merged1, df_products, on='_ProductID', how='left')
```

```python
df_merged2.head(10)
```

```python
df_merged2.info()
```

```python
#удаляем столбец, содержащий один и тот же параметр для всех строк
df_merged2 = df_merged2.drop(columns=['CurrencyCode'])
```

```python
#в столбцах, информация в которых имеет признак категориальных данных, меняем тип данных на category для облегчения группировок в дальнейшем
cat_cols = [ 'Sales Channel', 'WarehouseCode', '_SalesTeamID', '_CustomerID', '_StoreID', '_ProductID', 'Customer Names', 'Sales Team', 'Region', 'Product Name']
df_merged2[cat_cols] = df_merged2[cat_cols].astype('category')
```

```python
#заводим столбец для расчета финальной цены за единицу товара, после применения скидок
df_merged2['Final Unit Price'] = df_merged2['Unit Price'] * (1 - df['Discount Applied'])
```

```python
#заводим столбец "Выручка"
df_merged2['Revenue'] = df_merged2['Order Quantity'] * df_merged2['Final Unit Price']
```

```python
#заводим столбец "Маржа"
df_merged2['Margin'] = (df_merged2['Revenue'] - (df_merged2['Order Quantity'] * df_merged2['Unit Cost']))
```

```python
#заводим столбец "%Маржи"
df_merged2['%Margin'] = df_merged2['Margin']/df_merged2['Revenue']
```

```python
#считаем базовые статистики для числовых столбцов
df_merged2[['Order Quantity', 'Discount Applied','Unit Cost',  'Unit Price', 'Final Unit Price', 'Revenue']].describe()
```

```python
#Выбросов и аномалий в данных нет.
```

```python
#проверяем категориальные данные
for column in cat_cols:
  unique_values = df_merged2[column].unique()
  print(f"Уникальные значения в столбце {column}: {unique_values}")
  print("\n")
```

```python
# 1. Общая динамика продаж по времени
# -----------------------------
sales_over_time = df_merged2.groupby(pd.Grouper(key='OrderDate', freq='ME'))['Revenue'].sum()

plt.figure(figsize=(8,4))
sales_over_time.plot(kind='line', marker='*', color='purple')
plt.title('Общая динамика продаж за всё время')
plt.xlabel('Дата')
plt.ylabel('Выручка')
plt.grid(True)
plt.show()
```

```python
# Топ-5 команд по продажам
top_SalesTeam = df_merged2.groupby('Sales Team', observed=True)['Revenue'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(8,3))
top_SalesTeam.plot(kind='bar', color='purple')
plt.title('Топ-5 команд продаж по выручке')
plt.ylabel('Выручка')
plt.xlabel('Команда')
plt.show()
```

```python
# 2. Продажи по каналам
sales_by_channel = df_merged2.groupby('Sales Channel', observed=True)['Revenue'].sum()

plt.figure(figsize=(6,3))
sales_by_channel.plot(kind='bar', color='purple')
plt.title('Продажи по каналам')
plt.xlabel('Канал продаж')
plt.ylabel('Выручка')
plt.show()
```

```python
#Продажи по регионам
sales_by_channel = df_merged2.groupby('Region', observed=True)['Revenue'].sum()

plt.figure(figsize=(6,3))
sales_by_channel.plot(kind='bar', color='purple')
plt.title('Продажи по регионам')
plt.xlabel('Регион')
plt.ylabel('Выручка')
plt.show()
```

```python
# 3. Анализ сезонности (по месяцам)
df["Month"] = df["OrderDate"].dt.month
seasonality = df.groupby("Month")["Revenue"].sum()

plt.figure(figsize=(8,4))
seasonality.plot(kind='bar', color='purple')
plt.title('Сезонность продаж (по месяцам)')
plt.xlabel('Месяц')
plt.ylabel('Выручка')
plt.show()
```

🔹 Основные выводы

Unit Price ↔ Unit Cost (0.94)
Очень сильная положительная корреляция.
👉 Логично: чем дороже продукт, тем выше его себестоимость.

Revenue ↔ Unit Price (0.75) и Revenue ↔ Unit Cost (0.71)
Сильная связь: более дорогие товары (и с точки зрения цены, и себестоимости) дают больше выручки.

Revenue ↔ Order Quantity (0.52)
Средняя положительная корреляция: чем больше заказано единиц товара, тем выше выручка.
👉 Но не так сильно, как цена, т.е. дорогие товары важнее для общей выручки, чем количество.

Margin ↔ Revenue (0.79)
Очень сильная корреляция: высокая выручка почти всегда ведёт к высокой марже.
👉 Здесь можно дополнительно проанализировать продукты с большой выручкой, но маленькой маржой.

Margin ↔ Unit Price (0.57)
Связь есть, но не такая сильная, как с выручкой. Это значит, что высокая цена не всегда гарантирует высокую прибыльность (возможно, из-за себестоимости).

Discount Applied ↔ Revenue (–0.09) и ↔ Margin (–0.25)
Негативная слабая корреляция.
👉 Скидки слегка уменьшают маржу, но сильного эффекта на общую выручку почти нет.

Order Quantity ↔ Discount Applied (–0.01)
Связи почти нет: клиенты не покупают больше единиц товара только из-за скидки.

📌 Итог

Главный драйвер выручки — цена товара, а не количество.

Маржа почти полностью повторяет тренд выручки.


Скидки не стимулируют рост количества заказов, но немного «съедают» маржу.

```python
#выводим топ-5 продуктов с самой низкой маржинальностью
lowest_margin_products = (df_merged2.groupby('Product Name', observed=True)['%Margin'].mean().sort_values(ascending=True).head(5))  # сортируем по возрастанию
plt.figure(figsize=(6, 3))
lowest_margin_products.plot(kind='bar', color='purple')
plt.title('Топ-5 товаров с самой низкой маржинальностью (%)')
plt.ylabel('Маржинальность %')
plt.xlabel('Название товара')
plt.show()
```

**Нужно проверить, завист ли маржинальность от того, по какому каналу идут продажи**

Проверка поведения товаров по маржинальности в зависимости от канала продаж. Анализируется, отличается ли маржинальность товаров в зависимости от канала продаж (например, онлайн, офлайн, партнёрские продажи) по ключевым метрикам: маржа, выручка, доля скидки.
Для всех статистических тестов установлен уровень значимости 5% (α = 0.05).

Формулировка гипотез:
*   H₀ (нулевая гипотеза): Средняя маржинальность товаров одинакова для продаж он-лайн и в магазинах.

*   H₁ (альтернативная гипотеза): Средняя маржинальность товаров различается для продаж он-лайн и в магазинах.

t-статистика: -0.624, p-значение: 0.533


---

**Cтатистически значимых различий в маржинальности между каналами продаж не выявлено**.

```python
online_margin = df_merged2[df_merged2['Sales Channel'] == 'Online']['%Margin'].dropna()
store_margin = df_merged2[df_merged2['Sales Channel'] == 'In-Store']['%Margin'].dropna()
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(online_margin, store_margin, equal_var=False)
print(f't-статистика: {t_stat:.3f}, p-значение: {p_value:.3f}')
print('Cтатистически значимых различий в маржинальности между каналами продаж не выявлено.')
```
